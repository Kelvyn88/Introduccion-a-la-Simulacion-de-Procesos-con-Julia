```julia
# Problema 14.3
using JuMP
using Ipopt

# Define nombre del modelo
a = Model(solver=IpoptSolver(print_level=1))

# Define variables a emplear y sus límites
@variable(a, 1e-6 <= x1 <= 2000, start = 1745)
@variable(a, 1e-6 <= x2 <= 16000, start = 12000)
@variable(a, 1e-6 <= x3 <= 120, start = 110)
@variable(a, 1e-6 <= x4 <= 5000, start = 3048)
@variable(a, 1e-6 <= x5 <= 2000, start = 1974)
@variable(a, 85 <= x6 <= 93, start = 89.2)
@variable(a, 90 <= x7 <= 95, start = 92.8)
@variable(a, 3 <= x8 <= 12, start = 8)
@variable(a, 0.01 <= x9 <= 4, start = 3.6)
@variable(a, 145 <= x10 <= 162, start = 145)

# Define parámetros de desviación
d4l = 99/100
d4u = 100/99
d7l = 99/100
d7u = 100/99
d9l = 9/10
d9u = 10/9
d10l = 99/100
d10u = 100/99

# Define restricciónes
@NLconstraint(a, x1*(1.12 + 0.13167*x8 - 0.00667*x8^2) - d4l*x4 >= 0)
@NLconstraint(a, -(x1*(1.12 + 0.13167*x8 - 0.00667*x8^2)) + d4u*x4 >= 0)
@NLconstraint(a, (86.35 + 1.098*x8 - 0.038*x8^2 + 0.325*(x6 - 89)) - d7l*x7 >= 0)
@NLconstraint(a, -(86.35 + 1.098*x8 - 0.038*x8^2 + 0.325*(x6 - 89)) + d7u*x7 >= 0)
@NLconstraint(a, (35.82 - 0.222*x10) - d9l*x9 >= 0)
@NLconstraint(a, -(35.82 - 0.222*x10) + d9u*x9 >= 0)
@NLconstraint(a, (-133 + 3*x7) - d10l*x10 >=0)
@NLconstraint(a, -(-133 + 3*x7) + d10u*x10 >=0)

# Define igualdades
@NLconstraint(a, x5 + x1 - 1.22*x4 == 0)
@NLconstraint(a, ((98000*x3)/(x4*x9 + 1000*x3)) - x6 == 0)
@NLconstraint(a, ((x2 + x5)/x1) - x8 == 0)

# Define costos de operación
C1 = 0.063
C2 = 5.04
C3 = 0.035
C4 = 10
C5 = 3.36

# Define función objetivo
@NLobjective(a, Max, C1*x4*x7 - C2*x1 - C3*x2 - C4*x3 - C5*x5)

# Imprime modelo
print(a)

# Resuelve problema
solve(a)

print(@sprintf("Objective value: %2.6e\n", getobjectivevalue(a)))
print(@sprintf("x1: %2.6e\n", getvalue(x1)))
print(@sprintf("x2: %2.6e\n", getvalue(x2)))
print(@sprintf("x3: %2.6e\n", getvalue(x3)))
print(@sprintf("x4: %2.6e\n", getvalue(x4)))
print(@sprintf("x5: %2.6e\n", getvalue(x5)))
print(@sprintf("x6: %2.6e\n", getvalue(x6)))
print(@sprintf("x7: %2.6e\n", getvalue(x7)))
print(@sprintf("x8: %2.6e\n", getvalue(x8)))
print(@sprintf("x9: %2.6e\n", getvalue(x9)))
print(@sprintf("x10: %2.6e\n", getvalue(x10)))
```

```julia
# Problema 8.8

using JuMP
using Ipopt

m = Model(solver=IpoptSolver())

@variable(m, x1)
@variable(m, x2)

@constraint(m, x1^2 + x2^2 +x1 + x2 == 0)
@constraint(m, - x1 + x2^2 >= 0)

@NLobjective(m, Min, (x1 -1)^2 + x2^2)

solve(m)

println("X1 = ", getvalue(x1))
println("X2 = ", getvalue(x2))
```

```julia
# Problema 8.25

using JuMP
using Ipopt

m = Model(solver=IpoptSolver())

@variable(m, x[1:9] >= 0, start = 2)

@constraint(m, 1 - x[3]^2 - x[4]^2 >= 0)
@constraint(m, 1 - x[9]^2 >= 0)
@constraint(m, 1 - x[5]^2 - x[6]^2 >= 0)
@constraint(m, 1 - x[1]^2 - (x[2] - x[9])^2 >= 0)
@constraint(m, 1 - (x[1] - x[5])^2 - (x[2] - x[6])^2 >= 0)
@constraint(m, 1 - (x[1] - x[7])^2 - (x[2] - x[8])^2 >= 0)
@constraint(m, 1 - (x[3] - x[5])^2 - (x[4] - x[6])^2 >= 0)
@constraint(m, 1 - (x[3] - x[7])^2 - (x[4] - x[8])^2 >= 0)
@constraint(m, 1 - x[7]^2 - (x[8] - x[9])^2 >= 0)
@constraint(m, x[1]*x[4] - x[2]*x[3] >= 0)
@constraint(m, x[3]*x[9] >= 0)
@constraint(m, -x[5]*x[9] >= 0)
@constraint(m, x[5]*x[8] - x[6]*x[7] >= 0)

@NLobjective(m, Max, 0.5*(x[1]*x[4] - x[2]*x[3] + x[3]*x[9] - x[5]*x[9] + x[5]*x[8] - x[6]*x[7]))

solve(m)

println("f(x) = ", getobjectivevalue(m))

for i=1:9
    println("x$(i) = ", getvalue(x[i]))
end
```
