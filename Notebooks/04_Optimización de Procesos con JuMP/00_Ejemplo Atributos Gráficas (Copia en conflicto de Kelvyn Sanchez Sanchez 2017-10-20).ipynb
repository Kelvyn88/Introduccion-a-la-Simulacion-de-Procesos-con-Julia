```julia
n = 100
ts = linspace(0,8π,n)
x = ts .* map(cos,ts)
y = (0.1ts) .* map(sin,ts)
z = 1:n
plot(x,y,z,zcolor=reverse(z),m=(10,0.8,:blues,stroke(0)),leg=false,cbar=true,w=5)
plot!(zeros(n),zeros(n),1:n,w=10)
```

```julia
l = @layout([[a;b] c])
p = plot(plot([sin,cos],1,leg=false),scatter([atan,cos],1,leg=false),plot(log,1,xlims=(1,10π),ylims=(0,5),leg=false),layout=l)
display(p)
anim = Animation()
for x = linspace(1,10π,100)
    plot(push!(p,x,Float64[sin(x),cos(x),atan(x),cos(x),log(x)]))
    frame(anim)
end
```
