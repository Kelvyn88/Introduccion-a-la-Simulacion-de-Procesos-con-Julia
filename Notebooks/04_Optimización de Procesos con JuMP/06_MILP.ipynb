```julia
# Problema 16.2

using JuMP
using Cbc

# Define nombre del modelo
a = Model(solver = CbcSolver())

# Define variables a emplear y sus límites
#Variables continuas
# C[producto, unidad]
# @variable(a, C[1:4,1:3] >= 0)

@variable(a, C11 >= 0)
@variable(a, C12 >= 0)
@variable(a, C13 >= 0)
@variable(a, C21 >= 0)
@variable(a, C22 >= 0)
@variable(a, C23 >= 0)
@variable(a, C31 >= 0)
@variable(a, C32 >= 0)
@variable(a, C33 >= 0)
@variable(a, C41 >= 0)
@variable(a, C42 >= 0)
@variable(a, C43 >= 0)


#Variables binarias
#X[producto, posición]
#@variable(a, X[1:4,1:4], Bin)
@variable(a, X11, Bin)
@variable(a, X12, Bin)
@variable(a, X13, Bin)
@variable(a, X14, Bin)
@variable(a, X21, Bin)
@variable(a, X22, Bin)
@variable(a, X23, Bin)
@variable(a, X24, Bin)
@variable(a, X31, Bin)
@variable(a, X32, Bin)
@variable(a, X33, Bin)
@variable(a, X34, Bin)
@variable(a, X41, Bin)
@variable(a, X42, Bin)
@variable(a, X43, Bin)
@variable(a, X44, Bin)

C01 = 0
C02 = 0
C03 = 0
C04 = 0

#t = [3.5 1 1 1;
#3 4 5 6;
#5 6 7 8]

# t[product, unit]
t11 = 3.5
t12 = 4.3
t13 = 8.7
t21 = 4.0
t22 = 5.5
t23 = 3.5
t31 = 3.5
t32 = 7.5
t33 = 6.0
t41 = 12.0
t42 = 3.5
t43 = 8.0

# Define restricciónes lineales
@constraint(a, C11 >= C02)
@constraint(a, C21 >= C12)
@constraint(a, C31 >= C22)
@constraint(a, C41 >= C32)
@constraint(a, C12 >= C03)
@constraint(a, C22 >= C13)
@constraint(a, C32 >= C23)
@constraint(a, C42 >= C33)

# Define restricciónes que aseguran la posicion de solamente 1 producto en una
#secuencia
@constraint(a, X11 + X21 + X31 + X41 == 1)
@constraint(a, X12 + X22 + X32 + X42 == 1)
@constraint(a, X13 + X23 + X33 + X43 == 1)
@constraint(a, X14 + X24 + X34 + X44 == 1)

# Define restricciónes que aseguran que el producto este solamente en 1
#secuencia
@constraint(a, X11 + X12 + X13 + X14 == 1)
@constraint(a, X21 + X22 + X23 + X24 == 1)
@constraint(a, X31 + X32 + X33 + X34 == 1)
@constraint(a, X41 + X42 + X43 + X44 == 1)

# Define restricciónes g
# k=2
@constraint(a, C12 >= C11 + (X11*t12 + X21*t22 + X31*t32 + X41*t42))
@constraint(a, C22 >= C21 + (X12*t12 + X22*t22 + X32*t32 + X42*t42))
@constraint(a, C32 >= C31 + (X13*t12 + X23*t22 + X33*t32 + X43*t42))
@constraint(a, C42 >= C41 + (X14*t12 + X24*t22 + X34*t32 + X44*t42))

# k=3
@constraint(a, C13 >= C12 + (X11*t13 + X21*t23 + X31*t33 + X41*t43))
@constraint(a, C23 >= C22 + (X12*t13 + X22*t23 + X32*t33 + X42*t43))
@constraint(a, C33 >= C32 + (X13*t13 + X23*t23 + X33*t33 + X43*t43))
@constraint(a, C43 >= C42 + (X14*t13 + X24*t23 + X34*t33 + X44*t43))

# Define restricciones h
@constraint(a, C11 >= C01 + (X11*t11 + X21*t21 + X31*t31 + X41*t41))
@constraint(a, C21 >= C11 + (X12*t12 + X22*t22 + X32*t32 + X42*t42))
@constraint(a, C31 >= C21 + (X13*t13 + X23*t23 + X33*t33 + X43*t43))

# Define función objetivo
@objective(a, Min, C43)

# Imprime modelo
println(a)

status = solve(a)

println("Objective value: ", getobjectivevalue(a))

println(@sprintf("%i %i %i %i", getvalue(X11), getvalue(X21), getvalue(X31), getvalue(X41)))
println(@sprintf("%i %i %i %i", getvalue(X12), getvalue(X22), getvalue(X32), getvalue(X42)))
println(@sprintf("%i %i %i %i", getvalue(X13), getvalue(X23), getvalue(X33), getvalue(X43)))
println(@sprintf("%i %i %i %i", getvalue(X14), getvalue(X24), getvalue(X34), getvalue(X44)))
```
